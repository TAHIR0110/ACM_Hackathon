```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report

target_column = 'y'

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)

cat_model = CatBoostClassifier(iterations=100, random_state=42, verbose=0)
cat_model.fit(X_train, y_train, cat_features=np.where(X_train.dtypes != float)[0])

y_pred = cat_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=np.unique(y_test))

print(f"Classification Report:\n{report}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.metrics import accuracy_score, classification_report

target_column = 'y'

X_train_cat, X_test_cat, y_train_cat, y_test_cat = train_test_split(df, y, test_size=0.2, random_state=42)
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(df, y, test_size=0.2, random_state=42)

cat_model = CatBoostClassifier(iterations=100, random_state=42, verbose=0)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

voting_classifier = VotingClassifier(
    estimators=[
        ('cat', cat_model),
        ('rf', rf_model)
    ],
    voting='soft' 
)

voting_classifier.fit(X_train_cat, y_train_cat)

y_pred_soft = voting_classifier.predict(X_test_cat)

accuracy_soft = accuracy_score(y_test_cat, y_pred_soft)
report_soft = classification_report(y_test_cat, y_pred_soft, target_names=np.unique(y_test_cat))

print(f"Classification Report (Soft Voting):\n{report_soft}")

```
